```python
# Read marketing data

df = pd.read_excel(r'C:\Users\zdtra\OneDrive\Documents\Data\marketing_campaign.xlsx', index_col=0)
df.head(5)
```

```python
# Correlation Matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# Heat Map
plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix)
```

```python
# Assigning to X and y

X = df.drop(['Response','Education', 'Marital_Status','Dt_Customer'], axis=1)
print(X)
```

```python
y = df.iloc[:, -1].values
print(y)
```

```python
# Splitting to x and y
from sklearn.model_selection import train_test_split

X_Train, X_Test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(X_Train, y_train)
```

```python
y_pred = knn.predict(X_Test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print("The accuracy: {}".format(acc))
```
